```python
pd.crosstab(index=df["Product"],columns=df['Gender'],margins=True,normalize='index').T*100
```

KP281 was equally preferred by male and female.

KP481 was preferred by male followed closely by female.

KP781 was preferred mostly by male and was least preferred by females.
